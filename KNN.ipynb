```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from collections import Counter
```

```python
class KNN:
    
    def __init__(self,K=3):
        self.K=K 
        self.X=None
        self.Y=None

    def fit(self,X,Y):
        self.X=X
        self.Y=Y

    def distance(self,x,y):
        return np.linalg.norm(x-y)
    
    def count(self,n):
        votes=[]

        for i in n:
            votes.append(self.Y[i[0]])
        votes=Counter(votes)
        return votes.most_common()[0][0]

    def predict(self,x):
        y_pred=[]
        
        for i in x:
            dist=[]
            for j in self.X:
                dist.append(self.distance(i,j))
                neighbour=sorted(list(enumerate(dist)),key=lambda x:x[1])[0:self.K]
                label=self.count(neighbour)
                y_pred.append(label)
                return np.array(y_pred)
```
