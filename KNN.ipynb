```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,accuracy_score
from sklearn.model_selection import train_test_split as tt
import matplotlib.pyplot as plt
```

```python
class KNN:
    
    def __init__(self,K=3):
        self.K=K 
        self.X=None
        self.Y=None

    def fit(self,X,Y):
        self.X=X
        self.Y=Y

    def distance(self,x,y):
        return np.linalg.norm(x-y)
    
    def count(self,n):
        votes=[]

        for i in n:
            votes.append(self.Y[i[0]])
        votes=Counter(votes)
        return votes.most_common()[0][0]

    def predict(self,x):
        y_pred=[]
        
        for i in x:
            dist=[]
            for j in self.X:
                dist.append(self.distance(i,j))
            neighbour=sorted(list(enumerate(dist)),key=lambda x:x[1])[0:self.K]
            label=self.count(neighbour)
            y_pred.append(label)
        return np.array(y_pred)
```

```python
df=pd.read_csv('Iris.csv')
le=LabelEncoder()
df.drop('Id',axis=1)
df=np.array(df)
X=df[:,1:5]
Y=le.fit_transform(df[:,5])
x_train,x_test,y_train,y_test=tt(X,Y,test_size=0.2,random_state=2)
```

```python
knn=KNN(K=15)
knn.fit(x_train,y_train)
knn.predict(x_test)
```

```python
N=1
maxacc=0
for i in range(1,int(np.sqrt(x_train.shape[0])),2):
    knnM=KNN(K=i)
    knnM.fit(x_train,y_train)
    acc=accuracy_score(y_test,knnM.predict(x_test))
    if acc>maxacc:
        N=i
        maxacc=acc
```

```python
model=KNeighborsClassifier()
model.fit(x_train,y_train)
model.predict(x_test)
```

```python
print(classification_report(knn.predict(x_test),y_test))
print(classification_report(model.predict(x_test),y_test))
```
